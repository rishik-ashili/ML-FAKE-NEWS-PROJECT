The following is the code for preprocessing using TreeBankWord Tokenizer  
Aim of this part was to remove all unwanted characters and urls  
We also remove stopwords which are of least meaning and increasing understanding of text

NOTE : Due to lack of resources , the following code was ran on a cloud and output is used in this notebook

## PARTS OF SPEECH


The following is the code for Parts of Speech Tagging  
this can help to increase the understanding of the text by specifying what all the text meaning grammatically  
We have used Averaged Perceptron tagger for english data  
We have used Stanford's Nlp library Stanza which gives support for hindi POS Tagging  
the following code was ran on a cloud and output is used in this code


NOTE : Due to lack of resources , the following code was ran on a cloud and output is used in this notebook


```python
import pandas as pd
import nltk
from nltk import pos_tag
import re
import stanza

# Downloading necessary resources
nltk.download('averaged_perceptron_tagger')
stanza.download('hi')  # Downloaded the Hindi model

# Initializing the Stanza pipeline for Hindi with GPU
try:
    nlp_hi = stanza.Pipeline('hi', processors='tokenize,pos', use_gpu=True)  # Set use_gpu=True to use GPU
except Exception as e:
    print("Error initializing Stanza pipeline:", e)
    nlp_hi = stanza.Pipeline('hi', processors='tokenize,pos', use_gpu=False)  # Fallback to CPU if GPU fails

HindEng_Pos = combined_df 

# A function to check if text is predominantly in English
def is_english(text):
    if isinstance(text, str):  # Check if the input is a string
        english_chars = re.findall(r'[a-zA-Z]', text)
        non_english_chars = re.findall(r'[^\x00-\x7F]', text)
        return len(english_chars) > len(non_english_chars)
    return False 

# This is a custom tokenizer that uses regex to extract english words
def custom_tokenizer(text):
    return re.findall(r'\b\w+\b', text)

# This is a function for POS tagging English sentences
def pos_tag_english(text):
    tokens = custom_tokenizer(text)
    return pos_tag(tokens)

# This is  a function for POS tagging Hindi sentences using Stanza
def pos_tag_hindi(text):
    doc = nlp_hi(text)
    return [(word.text, word.xpos) for sentence in doc.sentences for word in sentence.words]

# Applying the correct POS tagging function based on the language
def pos_tag_mixed_language(text):
    if isinstance(text, str) and text.strip():  
        if is_english(text):
            return pos_tag_english(text)
        else:
            return pos_tag_hindi(text)
    return None 

# Converting all values in the 'text' column to strings and handling all NaN values
HindEng_Pos['text'] = HindEng_Pos['text'].astype(str).fillna('')

# Applying POS tagging based on detected language
HindEng_Pos['pos_tags'] = HindEng_Pos['text'].apply(pos_tag_mixed_language)
HindEng_Pos
```

## LEMMATIZING


This is the code for lemmatizing Data  
Lemmatizing is better than Stemming as it keeps the meaning of the word intact while converting to base root  
Again we have used Stanford's Stanza for hindi dataset as it provides lemmatization support too  
We have used Langid to identify both the text present in the data


NOTE : Due to lack of resources , the following code was ran on a cloud and output is used in this notebook


```python
import stanza
import langid
import pandas as pd

# filling NaN with an empty string
HindEng_Pos['processed_text'] = HindEng_Pos['processed_text'].fillna("").astype(str)

# Downloading English and Hindi models 
stanza.download('en')
stanza.download('hi')

# Setting up pipelines with GPU 
nlp_en = stanza.Pipeline(lang='en', use_gpu=True)
nlp_hi = stanza.Pipeline(lang='hi', use_gpu=True)

# This is to lemmatize text based on language detection
def lemmatize_text(text, lang_pipeline):
    doc = lang_pipeline(text)
    lemmatized_tokens = [word.lemma for sentence in doc.sentences for word in sentence.words]
    return " ".join(lemmatized_tokens)

# This is to apply lemmatization based on detected language
def apply_lemmatization(text):
    lang, _ = langid.classify(text)
    if lang == 'en':
        return lemmatize_text(text, nlp_en)
    elif lang == 'hi':
        return lemmatize_text(text, nlp_hi)
    else:
        return text

# Processing data in batches for better memory management  
batch_size = 100  
lemmatized_texts = []

for start in range(0, len(HindEng_Pos), batch_size):
    batch = HindEng_Pos['processed_text'][start:start + batch_size]
    # Applying lemmatization
    lemmatized_batch = batch.apply(apply_lemmatization)
    lemmatized_texts.extend(lemmatized_batch)

# Adding lemmatized results to the  main DataFrame
HindEng_Pos['lemmatized_text'] = lemmatized_texts
```

Following is the code for translating english news into hindi  
We tried using Deep Translator which has Google Translator which gives us free access to google apis  
As this api has a max word limit so we use facebook's transformer to summarise the news into lesser words then translate  
We use multiprocessing to speed up


NOTE : Due to lack of resources , the following code was ran on a cloud and output is used in this notebook


```python
#TRANSLATE DATA 
import pandas as pd
from transformers import pipeline
from deep_translator import GoogleTranslator
from multiprocessing import Pool, cpu_count
import os


# Initializing Summarizer
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=0)  # Use GPU/TPU if available

# Initializing Translator
translator = GoogleTranslator(source='en', target='hi')

# Directory for periodic saves
os.makedirs("progress", exist_ok=True)

# This is  to summarize text
def summarize_text(text, max_length=200, min_length=100):
    try:
        summary = summarizer(text, max_length=max_length, min_length=min_length, do_sample=False)
        return summary[0]['summary_text']
    except Exception as e:
        print(f"Error summarizing text: {text[:50]}... | Error: {e}")
        return text  

# This is  to process text (summarize + translate)
def process_text(text):
    try:
        
        if len(text) > 4500:
            chunks = [text[i:i + 4500] for i in range(0, len(text), 4500)]
            summarized_text = ' '.join([summarize_text(chunk) for chunk in chunks])
        else:
            summarized_text = summarize_text(text)

        
        translated = translator.translate(summarized_text)
        return translated
    except Exception as e:
        print(f"Error processing text: {text[:50]}... | Error: {e}")
        return text  # Return original text if any step fails

# This is Parallel processing function
def parallel_process_texts(texts, num_workers=None):
    if num_workers is None:
        num_workers = cpu_count()  # Use all available CPUs
    with Pool(num_workers) as pool:
        return pool.map(process_text, texts)

# This is to Save 
def save_progress(df, start_index, batch_size):
    batch_df = df[start_index:start_index + batch_size]
    batch_df['text_hindi'] = parallel_process_texts(batch_df['text'].tolist())
    save_path = f"progress/translated_batch_{start_index}_{start_index + batch_size}.csv"
    batch_df.to_csv(save_path, index=False)
    print(f"Saved batch {start_index} to {start_index + batch_size} at {save_path}")
    return batch_df

# This is processing loop   
def process_and_save_batches(df, batch_size=200):
    completed_dfs = []
    for start in range(0, len(df), batch_size):
        completed_dfs.append(save_progress(df, start, batch_size))
    # Combine all batches into a final result
    final_df = pd.concat(completed_dfs, ignore_index=True)
    final_df.to_csv('summarized_translated_text_final.csv', index=False)
    print("Final file saved as 'summarized_translated_text_final.csv'")
    return final_df

# Running the processing
batch_size = 200  # Save every 200 rows
df = process_and_save_batches(df, batch_size=batch_size)
```

```python
import numpy as np 
import pandas as pd 
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader, TensorDataset

import torch
from transformers import BertTokenizer, BertModel
import torch.nn as nn

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('/kaggle/input/hindienng/Final_Hindi_English_Text.csv')
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
model = BertModel.from_pretrained('bert-base-multilingual-cased').to(device)
```

```python
df['lemmatized_text'] = df['lemmatized_text'].astype(str)
tokens = df['lemmatized_text'].apply(lambda x: tokenizer(
    x,
    padding='max_length',
    truncation=True,
    max_length=512,
    return_tensors='pt'
))

df['input_ids'] = tokens.apply(lambda x: x['input_ids'][0].to(device))
df['attention_mask'] = tokens.apply(lambda x: x['attention_mask'][0].to(device))

def predict(model, input_ids, attention_mask):
    model.eval()
    with torch.o_grad():
        logits = model(input_ids, attention_mask)
        probabilities = nn.Softmax(dim=1)(logits)  
    return probabilities
```

```python
import torch.nn as nn
class FakeNewsClassifier(nn.Module):
    def __init__(self, num_classes=2):
        super(FakeNewsClassifier, self).__init__()
        self.bert = model
        self.classifier = nn.Linear(self.bert.config.hidden_size, 512)
        self.fc_1 = nn.Linear(512 , 384)
        self.fc_2 = nn.Linear(384 , 2)
        self.soft_max = nn.Softmax(dim = 1)

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls_output = outputs.last_hidden_state[:, 0, :]  
        logits = self.classifier(cls_output)  
        out = self.fc_1(logits)
        out = self.fc_2(out)
        out = self.soft_max(out)
        return out

fake_news_classifier = FakeNewsClassifier().to(device)
# fake_news_classifier = torch.load("/kaggle/input/model/pytorch/default/1/Final model_full .pth" , map_location = "cuda" )
```

```python
# Prepare data for DataLoader
from torch.cuda.amp import autocast, GradScaler

scaler = GradScaler()

def prepare_data(df):
    input_ids = torch.stack(list(df['input_ids'].values))
    attention_masks = torch.stack(list(df['attention_mask'].values))
    labels = torch.tensor(df['label'].values)
    return TensorDataset(input_ids, attention_masks, labels)

# Training function
def train(model, dataloader, optimizer, criterion):
    model.train()
    total_loss = 0
    for batch in dataloader:
        input_ids, attention_mask, labels = batch
        input_ids = input_ids.to(device)
        attention_mask  = attention_mask.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        with autocast():
            logits = model(input_ids, attention_mask)
            loss = criterion(logits, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(dataloader)

# Evaluation function
def evaluate(model, dataloader, criterion):
    model.eval()
    total_loss, correct_predictions = 0, 0
    with torch.no_grad():
        for batch in dataloader:
            input_ids, attention_mask, labels = batch
            input_ids = input_ids.to(device)
            attention_mask  = attention_mask.to(device)
            labels = labels.to(device)
            with autocast():
                logits = model(input_ids, attention_mask)
                loss = criterion(logits, labels)
                
            total_loss += loss.item()
            preds = torch.argmax(logits, dim=1)
            correct_predictions += (preds == labels).sum().item()
    accuracy = correct_predictions / len(dataloader.dataset)
    return total_loss / len(dataloader), accuracy

# To make predictions with probabilities
def predict(model, input_ids, attention_mask):
    model.eval()
    with torch.no_grad():
        logits = model(input_ids, attention_mask)
        probabilities = nn.Softmax(dim=1)(logits)
    return probabilities
```

```python
optimizer = optim.Adam(model.parameters(), lr=1e-5)
criterion = nn.CrossEntropyLoss()
torch.cuda.empty_cache()
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)
train_data = prepare_data(train_df)
val_data = prepare_data(val_df)
train_loader = DataLoader(train_data, batch_size=8, shuffle=True)
val_loader = DataLoader(val_data, batch_size=8, shuffle=False)

epochs = 10
for epoch in range(epochs):
    train_loss = train(fake_news_classifier, train_loader, optimizer, criterion)
    val_loss, val_accuracy = evaluate(fake_news_classifier, val_loader, criterion)
    print(f"Epoch {epoch+1}/{epochs}")
    print(f"Train Loss: {train_loss:.4f}")
    print(f"Validation Loss: {val_loss:.4f}")
    print(f"Validation Accuracy: {val_accuracy:.4f}")
```
